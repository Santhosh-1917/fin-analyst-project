```python
import pandas as pd

# --- 1Ô∏è‚É£ Read the cleaned 24-column CSV ---
file_path = "../data/processed/apple_clean_24cols.csv"
df_clean24 = pd.read_csv(file_path)

# --- 2Ô∏è‚É£ Quick verification ---
print(f"‚úÖ Successfully loaded: {file_path}")
print(f"üìä Shape: {df_clean24.shape}")
print("\nüìÅ Columns:")
print(df_clean24.columns.tolist())

# --- 3Ô∏è‚É£ Optional: quick preview ---
display(df_clean24.head())
```

```python
# --- 1Ô∏è‚É£ Check for missing values ---
missing_summary = df_clean24.isna().sum().sort_values(ascending=False)

print("üîç Missing Values Summary:\n")
print(missing_summary)

# --- 2Ô∏è‚É£ Percentage of missing values ---
missing_percent = (df_clean24.isna().mean() * 100).round(1)
print("\nüìä Percentage of Missing Values:\n")
print(missing_percent)

# --- 3Ô∏è‚É£ Optional: visualize missing data pattern ---
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
missing_percent.plot(kind='bar', color='teal')
plt.title("Percentage of Missing Values per Column", fontsize=14)
plt.ylabel("Percentage (%)")
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

```python
# --- 1Ô∏è‚É£ Filter dataset for 2010‚Äì2025 ---
df_recent = df_clean24[df_clean24["Year"] >= 2010].copy()

# --- 2Ô∏è‚É£ Compute missing values within this window ---
missing_recent = df_recent.isna().sum()
missing_percent_recent = (df_recent.isna().mean() * 100).round(2)

# --- 3Ô∏è‚É£ Combine into a single summary table ---
missing_summary_recent = pd.DataFrame({
    "Missing_Count": missing_recent,
    "Missing_%": missing_percent_recent
}).sort_values(by="Missing_%", ascending=False)

print("üìÖ Missing Values (2010‚Äì2025 period only):")
display(missing_summary_recent)

# --- 4Ô∏è‚É£ Optional: visualize ---
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
missing_summary_recent["Missing_%"].plot(kind="bar", color="darkcyan")
plt.title("Percentage of Missing Values (2010‚Äì2025 Window)", fontsize=14)
plt.ylabel("Percentage (%)")
plt.xticks(rotation=75, ha='right')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# --- 1Ô∏è‚É£ Select key financial columns to visualize ---
cols_to_plot = [
    "IncomeLossFromContinuingOperationsBeforeIncomeTaxesExtraordinaryItemsNoncontrollingInterest",
    "NetCashProvidedByUsedInFinancingActivities",
    "NetCashProvidedByUsedInInvestingActivities",
    "NetCashProvidedByUsedInOperatingActivities"
]

# --- 2Ô∏è‚É£ Ensure sorted by date ---
df_clean24 = df_clean24.sort_values("end")

# --- 3Ô∏è‚É£ Plot trends ---
plt.figure(figsize=(12,6))
for col in cols_to_plot:
    plt.plot(df_clean24["end"], df_clean24[col], label=col, linewidth=2)

plt.title("üìà Key Financial Trends (2010‚Äì2025 Window)", fontsize=14)
plt.xlabel("Date")
plt.ylabel("Value (USD)")
plt.legend(loc="upper left", fontsize=8)
plt.grid(True, linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# Filter for analysis window
df_fill = df_clean24[df_clean24['Year'].between(2010, 2025)].copy()
```

```python
# Check missing percentages again
missing_summary = (df_fill.isna().sum() / len(df_fill)) * 100
cols_to_fill = missing_summary[missing_summary.between(0.1, 15)].index.tolist()

print("Columns to fill:", cols_to_fill)
```

```python
# Sort by date for correct interpolation
df_fill = df_fill.sort_values("end")

# Apply linear interpolation
df_fill[cols_to_fill] = df_fill[cols_to_fill].interpolate(method='linear')

# Optional: forward/backward fill any edge NaNs
df_fill[cols_to_fill] = df_fill[cols_to_fill].ffill().bfill()
```

```python
print("‚úÖ Missing values after interpolation:")
print(df_fill[cols_to_fill].isna().sum())
```
