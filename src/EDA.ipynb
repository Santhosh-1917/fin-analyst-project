```python
import pandas as pd

# Load the file
df_rev = pd.read_csv("../data/processed/apple_revenue_tags.csv")

# Basic info
print("üìÑ Shape of dataset:", df_rev.shape)
print("üìù Columns:", df_rev.columns.tolist())

# Convert 'end' to datetime for consistency
df_rev['end'] = pd.to_datetime(df_rev['end'], errors='coerce')

# Check for duplicate rows (full row duplicates)
full_dupes = df_rev[df_rev.duplicated()]
print(f"üîÅ Number of full duplicate rows: {len(full_dupes)}")
if len(full_dupes) > 0:
    display(full_dupes)

# Check for duplicate dates (same 'end' with different values)
date_dupes = df_rev[df_rev.duplicated(subset=['end'], keep=False)].sort_values('end')
print(f"üìÜ Number of duplicate 'end' dates: {date_dupes['end'].nunique()}")
if not date_dupes.empty:
    display(date_dupes)
```

```python
# Remove exact duplicates
df_clean = df_rev.drop_duplicates()

print("After dropping duplicates:", df_clean.shape)

# Optional: Check for duplicate dates (e.g. multiple revenue entries on the same 'end' date)
dup_dates = df_clean[df_clean.duplicated(subset=['end'], keep=False)]
print("Duplicate dates shape:", dup_dates.shape)
dup_dates.head()
```

```python
# Suppose df_revenue is the cleaned 215-row revenue dataset
df_clean['end'] = pd.to_datetime(df_clean['end'])

# Option 1: Keep only the last reported value per 'end' date
df_revenue_agg = df_clean.groupby('end').max().reset_index()

print("Aggregated revenue dataset shape:", df_revenue_agg.shape)
```
